```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 6))
plt.hist(df['x1'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of x1')
plt.xlabel('x1')
plt.ylabel('Frequency')
plt.show()

y_counts = df['y'].value_counts()

plt.figure(figsize=(10, 6))
plt.bar(y_counts.index, y_counts.values, color='blue', edgecolor='black')
plt.title('Composition of y')
plt.xlabel('y')
plt.ylabel('Count')
plt.show()
```
